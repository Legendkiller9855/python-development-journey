```python
def rem(l , word):
  n = []
  for item in l:
    if not(item == word):
      n.append(item.strip(word))
  return n

l = ["Harry","Vivek","Rohan","an","krishanhai"]
print(rem(l,"an"))


# def rem(l, word):
# → Defines a function named rem.
# → It takes two parameters:
#    l = the list of strings
#    word = the word/string to remove

# n = []
# → Creates an empty list named n.
# → This list will store the final updated items.

# for item in l:
# → Starts a loop to go through each element of the list l.
# → item will take values one by one from the list.

# if not(item == word):
# → Checks whether the current item is NOT equal to word.
# → If item is exactly equal to word, it will be skipped.

# n.append(item.strip(word))
# → Adds a modified version of item into the list n.
# → strip(word) removes the characters present in 'word'
#    only from the beginning and end of the string.
# → Example:
#    "Rohan".strip("an") becomes "Roh"
#    because 'a' and 'n' are removed from the end.
# → If item is exactly equal to word, it is not added at all.

# return n
# → Returns the new list n after the loop is complete.

# l = ["Harry", "Vivek", "Rohan", "an"]
# → Creates a list of strings and stores it in variable l.

# print(rem(l, "an"))
# → Calls the rem() function with:
#    l as the list
#    "an" as the word
# → Prints the returned list.

# Final Output:
# → ['Harry', 'Vivek', 'Roh']
# → "an" is removed completely, and "Rohan" becomes "Roh".
```
