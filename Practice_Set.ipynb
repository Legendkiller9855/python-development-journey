```python
def rem(l , word):
  n = []
  for item in l:
    if not(item == word):
      n.append(item.strip(word))
  return n

l = ["Harry","Vivek","Rohan","an","krishanhai"]
print(rem(l,"an"))


# def rem(l, word):
# → Defines a function named rem.
# → It takes two parameters:
#    l = the list of strings
#    word = the word/string to remove

# n = []
# → Creates an empty list named n.
# → This list will store the final updated items.

# for item in l:
# → Starts a loop to go through each element of the list l.
# → item will take values one by one from the list.

# if not(item == word):
# → Checks whether the current item is NOT equal to word.
# → If item is exactly equal to word, it will be skipped.

# n.append(item.strip(word))
# → Adds a modified version of item into the list n.
# → strip(word) removes the characters present in 'word'
#    only from the beginning and end of the string.
# → Example:
#    "Rohan".strip("an") becomes "Roh"
#    because 'a' and 'n' are removed from the end.
# → If item is exactly equal to word, it is not added at all.

# return n
# → Returns the new list n after the loop is complete.

# l = ["Harry", "Vivek", "Rohan", "an"]
# → Creates a list of strings and stores it in variable l.

# print(rem(l, "an"))
# → Calls the rem() function with:
#    l as the list
#    "an" as the word
# → Prints the returned list.

# Final Output:
# → ['Harry', 'Vivek', 'Roh']
# → "an" is removed completely, and "Rohan" becomes "Roh".
```

```python
class Employees():
  # Class attributes are not strictly necessary if __init__ always sets instance attributes
  def __init__(self , name , city , language , salary):
    self.name = name
    self.city = city
    self.language = language
    self.salary = int(salary) # Convert salary to int as it's typically numeric

# Get user inputs first
Employee_name = input("What's your name?")
Employee_city = input("What's your city?")
Employee_language = input("What's your language?")
Employee_salary = input("What's your salary?")

# Then, create the object with the collected inputs
Employee_name1 = Employees(Employee_name , Employee_city , Employee_language , Employee_salary)
print(Employee_name1.name , Employee_name1.city , Employee_name1.language , Employee_name1.salary)
```

```python
number = float(input("What's your number?"))
class Calculator():
  def __init__(self ,number):
    self.number = number

  def square(self):
    print(f"The square of the number is : {self.number *self.number}")


  def cube(self):
    print(f"The cube of the number is : {self.number*self.number*self.number}")

  def squareroot(self):
    print(f"The squareroot of the number is : {self.number**1/2}")
a = Calculator(number)
a.squareroot()
a.square()
a.cube()
```

```python
from random import randint
class Train():
  def __init__(self, trainNo, fro, to):
    self.trainNo = trainNo
    self.fro = fro
    self.to = to

  def status(self):
    print(f"the train no. : {self.trainNo} is on time.")

  def fare(self):
    print(f"The train no. : {self.trainNo} running from {self.fro} to {self.to} has a fare costing : {randint(500,2001)}")

  def seats(self):
    print(f"The train no. : {self.trainNo} running from {self.fro} to {self.to} has {randint(23,89)} number of seats available ")

a = Train(144231,"delhi","Ludhiana")
a.status()
a.fare()
a.seats()
```
